# 📚 Matplotlib Tutorial – Drawing & Explaining Different Graphs

This notebook demonstrates how to use **Matplotlib** to create and customize different types of plots, charts, and graphs.

## **1. Introduction to Matplotlib**
**Matplotlib** is a Python library used for creating static, animated, and interactive visualizations. The `matplotlib.pyplot` module provides a MATLAB-like interface for creating plots.

````python
import matplotlib.pyplot as plt
import numpy as np
````

## **2. Basic Line Plot**
Line plots are used to show relationships between variables, often trends over time.


````python
x = [1, 2, 3, 4, 5]
y = [2, 4, 6, 8, 10]

plt.plot(x, y, color='red', marker='*', linestyle='--')

plt.title("Basic Line Plot")
plt.xlabel("X-axis")
plt.ylabel("Y-axis")
plt.grid(True)
plt.show()
````

## **3. Bar Chart**
Bar charts compare categories using rectangular bars.


````python
categories = ['A', 'B', 'C', 'D']
values = [10, 15, 7, 12]

# https://matplotlib.org/stable/users/explain/colors/colors.html

plt.bar(categories, values, color='green')
plt.title("Bar Chart Example")
plt.xlabel("Categories")
plt.ylabel("Values")
plt.show()
````

## **4. Histogram**
Histograms show the frequency distribution of data.


````python
data = np.random.randn(1000)
data
````

````python
plt.hist(data, bins=20, color='green', edgecolor='black')
plt.title("Histogram Example")
plt.xlabel("Bins")
plt.ylabel("Frequency")
plt.show()
````

## **5. Scatter Plot**
Scatter plots show relationships between two continuous variables.


````python
x = [5, 7, 8, 7, 6, 9, 5, 6, 7, 8]
y = [99, 86, 87, 88, 100, 86, 103, 87, 94, 78]

plt.scatter(x, y, color='red')
plt.title("Scatter Plot Example")
plt.xlabel("X-axis")
plt.ylabel("Y-axis")
plt.show()
````

## **6. Pie Chart**
Pie charts display proportions of a whole.


````python
sizes = [30, 20, 25, 5]
labels = ['Apples', 'Bananas', 'Cherries', 'Dates']
colors = ['gold', 'lightgreen', 'lightcoral', 'skyblue']

plt.pie(sizes, labels=labels, colors=colors, autopct='%1.1f%%', startangle=140)
plt.title("Pie Chart Example")
plt.show()
````

## **7. Multiple Plots in One Figure**
You can use `subplot()` to compare multiple plots side-by-side.


````python
x = [1, 2, 3, 4, 5]
y1 = [1, 4, 9, 16, 25]
y2 = [25, 16, 9, 4, 1]

plt.subplot(1, 2, 1)
plt.plot(x, y1, 'r-')
plt.title("Square Numbers")

plt.subplot(1, 2, 2)
plt.plot(x, y2, 'b-')
plt.title("Reverse Numbers")

plt.tight_layout()
plt.show()
````

## **8. Customizing Styles**
Matplotlib has built-in styles to change the look of your plots.


````python
plt.style.use('ggplot')
x = [1, 2, 3, 4, 5]
y = [2, 3, 5, 7, 11]
plt.plot(x, y, marker='o')
plt.title("Styled Line Plot")
plt.show()
plt.style.use('default')
````

## **9. Real-World Example – Comparing Multiple Graph Types**


````python
x = np.arange(1, 6)
y = np.array([2, 4, 6, 8, 10])

plt.figure(figsize=(10, 6))

plt.subplot(2, 2, 1)
plt.plot(x, y, 'o-g')
plt.title("Line Plot")

plt.subplot(2, 2, 2)
plt.bar(x, y, color='orange')
plt.title("Bar Chart")

plt.subplot(2, 2, 3)
plt.scatter(x, y, color='purple')
plt.title("Scatter Plot")

plt.subplot(2, 2, 4)
plt.pie(y, labels=[f'Item {i}' for i in x], autopct='%1.1f%%')
plt.title("Pie Chart")

plt.tight_layout()
plt.show()
````
