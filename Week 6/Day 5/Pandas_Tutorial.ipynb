````python
import pandas as pd
data = {
    "roll No": [221,222,223,224,225],
    "subj1":[50,60,70,80,90],
    "subj2":[90,80,70,60,50],
    "subj3":[95,85,75,65,55]
}
df= pd.DataFrame(data)
print(df)
````

````python
df.describe()
````

````python
df["subj1"].mean()
````

````python
df["avg_marks"] = (df['subj1'] + df['subj2'] + df['subj3'])/3
df
````

````python
df[['subj1', 'subj2', 'subj3']] # Selected 3 Subject Cols 
````

````python
df['Avg2'] = df[['subj1', 'subj2', 'subj3']].mean(axis=1) # axis = 1 >> row-wise avg
df
````
