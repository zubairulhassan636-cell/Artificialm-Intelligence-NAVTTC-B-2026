### Class Student

### Class Car

```python
class Car:
    # Class attributes
    company = "Toyota"
    wheels = 4

    # Constructor with 5 object attributes
    def __init__(self, model, color, year, price, mileage):
        self.model = model
        self.color = color
        self.year = year
        self.price = price
        self.mileage = mileage

    # Object Method 1
    def start_car(self):
        print(self.model, "is starting...")

    # Object Method 2
    def stop_car(self):
        print(self.model, "is stopping...")

    # Object Method 3
    def display_info(self):
        print("Model:", self.model)
        print("Color:", self.color)
        print("Year:", self.year)
        print("Price:", self.price)
        print("Mileage:", self.mileage)

    # Class Method
    @classmethod
    def show_company_info(cls):
        print("Company:", cls.company)
        print("Number of Wheels:", cls.wheels)


# Creating objects
car1 = Car("Corolla", "White", 2022, 4500000, "15 km/l")
car2 = Car("Civic", "Black", 2023, 5200000, "14 km/l")

# Calling object methods
car1.display_info()
car1.start_car()
car1.stop_car()

# Calling class method
Car.show_company_info()
```

### Class Animal

```python
class Animal:
    # Class Attributes
    kingdom = "Animalia"
    planet = "Earth"

    # Constructor (Object Attributes)
    def __init__(self, name, species, age, color, weight):
        self.name = name
        self.species = species
        self.age = age
        self.color = color
        self.weight = weight

    # Object Method 1
    def eat(self):
        print(self.name, "is eating.")

    # Object Method 2
    def sleep(self):
        print(self.name, "is sleeping.")

    # Object Method 3
    def display_info(self):
        print("Name:", self.name)
        print("Species:", self.species)
        print("Age:", self.age)
        print("Color:", self.color)
        print("Weight:", self.weight)

    # Class Method
    @classmethod
    def show_class_info(cls):
        print("Kingdom:", cls.kingdom)
        print("Planet:", cls.planet)


# Creating Objects
animal1 = Animal("Lion", "Mammal", 5, "Golden", 190)
animal2 = Animal("Parrot", "Bird", 2, "Green", 1)

# Calling Object Methods
animal1.display_info()
animal1.eat()
animal1.sleep()

print()

animal2.display_info()

print()

# Calling Class Method
Animal.show_class_info()
```
