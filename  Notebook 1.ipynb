```python
# Task 7
import cmath

a = 1
b = 5
c = 6

# Calculate the discriminant
d = (b**2) - (4*a*c)

# Find two solutions (works even if d is negative)
sol1 = (-b - cmath.sqrt(d)) / (2*a)
sol2 = (-b + cmath.sqrt(d)) / (2*a)

print(sol1, sol2)
```
