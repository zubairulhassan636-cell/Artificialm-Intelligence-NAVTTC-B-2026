### String Concatenation

```python
a = "Helllo"
b = "World"
print(a+" "+b)
```

```python
# Iterate string through for loop
greet= "Hello World"
for i in greet:
    print(i)
```

### String Membership 

```python
e = "Python Programming"
print('t' in e)
print('n' not in e)
```

### String Built-in Function

```python
e = "Python Programming"
print(e.upper())
```

```python
e = "Python Programming"
print(e.lower())
```

```python
e = "Python Programming"
print(e.partition(" "))
```

```python
e = "Python Programming"
print(e.split())
```
