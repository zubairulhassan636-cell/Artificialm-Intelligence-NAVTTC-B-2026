# Dictionary

## Dictionary Creation

```python
country_capitals = {
    "Germany" : "Berlin",
    "Canada" : "Ottawa",
    "England" : "London"}
print(country_capitals)
```

## Access Dictionary items

```python
country_capitals = {
    "Germany" : "Berlin",
    "Canada" : "Ottawa",
    "England" : "London"}
print(country_capitals["Germany"])
print(country_capitals["England"])
```

## Add item to Dictionary

```python
country_capitals["Italy"] = "Rome"
print(country_capitals)
```

## Remove Dictionary item

```python
country_capitals = {
    "Germany" : "Berlin",
    "Canada" : "Ottawa",
    "England" : "London"}
del country_capitals["Germany"]
print(country_capitals)
```

```python
country_capitals = {
    "Germany" : "Berlin",
    "Canada" : "Ottawa",
    "England" : "London"}
country_capitals.clear()
print(country_capitals)
```

## Change Dictionary item

```python
country_capitals = {
    "Germany" : "Berlin",
    "Canada" : "Ottawa",
    "Italy" : "Naples",
    "England" : "London"}
country_capitals["Italy"] = "Rome"
print(country_capitals)
```

## Iterating through Dictionary

```python
country_capitals = {
    "Germany" : "Berlin",
    "Canada" : "Ottawa",
    "Italy" : "Naples",
    "England" : "London"}
for i in country_capitals:
    print(i)
```

```python
country_capitals = {
    "Germany" : "Berlin",
    "Canada" : "Ottawa",
    "Italy" : "Rome",
    "England" : "London"}
for i in country_capitals:
    capital=country_capitals[i]
    print(capital)
```

## Dictionary length

```python
country_capitals = {
    "Germany" : "Berlin",
    "Canada" : "Ottawa",
    "Italy" : "Rome",
    "England" : "London"}
print(len(country_capitals))
```

## Dictionary Membership Test

```python
file_types = {
    ".pdf" : "PDF Document",
    ".txt" : "Text File",
    ".jpg" : "JPEG Image"}
print(".pdf" in file_types)
print(".mp3" in file_types)
print(".mp3" not in file_types)
```

### Task

```python
dict = {"name" : "Ali", "age" : 18, "city" : "Multan", "institute" : "Corvit"}
print(dict)
# Add new value pair
dict["country"] = "Pakistan"
print(dict)
# Modify age value
dict["age"] = 22
print(dict)
#Access value city
print(dict["city"])
# Remove last value pair
del dict["institute"]
print(dict)
# Check if city is present in dictionary
print("city" in dict)
```

```python
#Merge dictionary
numbers = {"ten" : 10, "twenty" : 20, "thirty" : 30}
num = {"fourty" : 40, "fifty" : 50, "sixty" : 60}
numbers.update(num)
print(numbers)
```

```python
string="NAVTTC"
freq={}
for char in string:
    if char in freq:
        freq[char]+=1
    else:
        freq[char]=1
print(freq)
```
