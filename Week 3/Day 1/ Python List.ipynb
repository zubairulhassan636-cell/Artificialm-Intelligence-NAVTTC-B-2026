```python
# Find unique values and count order
customer_order = ["Mango", "Dates", "Apple",
                  "pineApple", "Apple", "Dates",
                  "Mango", "Dates", "Apple"]

count_items = {}

for item in customer_order:
    if item in count_items:
        count_items[item] += 1
    else:
        count_items[item] = 1

print(count_items)

customer_order = set(customer_order)
print(customer_order)
```

```python
# Remove Duplicates
student_id = [11, 22, 33, 11, 22, 33, 44, 66, 66]
student_id = set(student_id)
print(student_id)
```

```python
# Remove Even values
student_id = [11, 22, 33, 11, 22, 33, 44, 66, 66]
odd_numbers = []
for n in student_id:
    if n % 2 != 0:
        odd_numbers.append(n)
print(odd_numbers)
```
