# Iterators

```python
# Iterating through an iterator
my_list = [3, 4, 6]
iterator = iter(my_list)

# Get the first element of iterator
print(next(iterator))

# Get the second element of iterator
print(next(iterator))

# Get the third element of iterator
print(next(iterator))
```

```python
# Using for loop
my_list = [3, 4, 6]
for element in my_list:
    print(element)
```

```python
my_list = [1, 2, 3, 4, 5, 6, 7, 8]
iterator = iter(my_list)
for element in iterator:
    print(element)
```

```python
# List iteration using iterators
a = [2, 3, 5, 8, 6]
iterator = iter(a)
print(next(iterator))
print(next(iterator))
print(next(iterator))
print(next(iterator))
print(next(iterator))
```

```python
# Tuple iteration using iterators
b = ("Hello", 23, 35, "Python")
iterator = iter(b)
print(next(iterator))
print(next(iterator))
print(next(iterator))
print(next(iterator))
```

```python
# Dictionary iteration using iterators
c = {"name" : "Ali", "age" : 18, "city" : "Multan"}
iterator = iter(c.items())
print(next(iterator))
print(next(iterator))
print(next(iterator))
```

```python
# set iteration using iterators
d = {45, "Java", 24, "abc"}
iterator = iter(d)
print(next(iterator))
print(next(iterator))
print(next(iterator))
print(next(iterator))
```

```python
# string iteration using iterators
e = "Python"
iterator = iter(e)
print(next(iterator))
print(next(iterator))
print(next(iterator))
print(next(iterator))
print(next(iterator))
print(next(iterator))
```
