# Python Lambda functions

```python
x = lambda a : a + 10
print(x(10))
```

```python
y = lambda a, b : a * b
print(y(5, 6))
```

```python
x = lambda a, b, c : a + b + c
print(x(5, 7, 8))
```

```python
def myfunc(n):
    return lambda a: a * n

double = myfunc(2)
print(double(5))
```

```python
check = lambda x : "Even" if x % 2==0 else "odd"
check(8)
```

# Map function

```python
sqr = lambda x : x**2
a = [2, 3, 4, 5, 6]
squared_value = map(sqr, a)
print(list(squared_value))
```

```python
cube = lambda y : y**3
b = {3, 4, 5, 6}
cube_value = map(cube, b)
print(set(cube_value))
```

```python
add = lambda a, b : a+b
a = [2, 4, 6, 8]
b = [3, 5, 7, 9]
add_value = map(add, a, b)
print(list(add_value))
```

# Filter function

```python
# Using lambda function
numbers = [1, 2, 3, 4, 5, 6, 7, 8]
even = filter(lambda x : x%2==0, numbers)
print(list(even))
```

```python
# Using lambda function
numbers = [1, 2, 3, 4, 5, 6, 7, 8]
odd = filter(lambda x : x%2 !=0, numbers)
print(list(odd))
```

```python
# Using simple function
def even(x):
    return x % 2 ==0
numbers = [1, 2, 3, 4, 5, 6, 7, 8]
result = filter(even, numbers)
print(list(result))
```
