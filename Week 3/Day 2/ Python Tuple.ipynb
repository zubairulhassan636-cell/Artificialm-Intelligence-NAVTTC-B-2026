# Access Tuple items

## Indexing

```python
tuple=(23, "Hello", 34, 'abc')
print(tuple[1])
```

```python
# Negative indexing
c=("Python", "Programming", 23, 35, "Hello")
print(c[-1])
```

```python
# Slicing
print(c[1:4])
```

```python
print(c[0:5:2])
```

```python
print(c[:4])
```

```python
# Check if item exist
tuple=("Apple", "mango", "Grapes", "Banana", "Orange")
if "Apple"  in tuple:
    print("Apple is present in tuple")
```

```python
# using for loop for slicing
e=(1,2,3,4,5,6,7,8)
for i in e:
    print(i)
```

```python
f=(1,2,2,3,4,5,2,6,7,)
f.count(2)
```

```python
f.index(5)
```

```python
# Join tuples
tuple1=(1,2,3,4)
tuple2=(5,6,7,8)
tuple3=tuple1+tuple2
print(tuple3)
```
