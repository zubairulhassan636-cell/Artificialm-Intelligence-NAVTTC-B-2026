# Set 

## Set Creation

```python
a = {1, 2, 4, 6, 5, 3}
print(a)
```

```python
number = {2, 4, 6, 2, 8}
print(number)
print(type(number))
```

## Add item in set

```python
numbers = {21, 34, 54, 12}
print("Initial set :",numbers)
numbers.add(32)
print("Updated set :",numbers)
```

## Update Set

```python
companies = {"lacoste", "RalphLauren"}
tech_companies = {"apple", "google", "apple"}
companies.update(tech_companies)
print(companies)
```

## Remove Element from set

```python
languages = {"Python", "C++", "Java"}
print("Inital set :", languages)
languages.remove("Java")
print("Set after removing element :", languages)
```

## Length of set

```python
a = {2, 3, 4, 5, 6, 7, 8}
print(len(a))
```
